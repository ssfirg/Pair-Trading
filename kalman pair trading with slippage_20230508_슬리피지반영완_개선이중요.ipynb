```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import FinanceDataReader as fdr


class BacktestStrategy:
    def __init__(self):
        pass

    # 가져온 데이터들로 total, units, cash 데이터프레임을 만들어줌
    def basic_tool(self, initial_amount = 10000000):
        price = self.price
        units = pd.DataFrame(index= price.index, columns = price.columns,  dtype=float).fillna(0)
        total = pd.Series(index = price.index,  dtype=float).fillna(0) # 전체 총자산을 나타내는
        cash = pd.Series(index = price.index ,  dtype=float).fillna(0)
        
        total.iloc[0] , cash.iloc[0] = initial_amount, initial_amount
        return total , units, cash

    def order(self, bar, weight_t0, weight_t1, units, cash):
        if np.array_equal(weight_t0, weight_t1):
            units.iloc[bar] = units.iloc[bar-1]
        else:
            units.iloc[bar] = ((units.iloc[bar - 1] @ self.price.iloc[bar] + cash.iloc[bar]) * np.array(weight_t0))/ self.price.iloc[bar]
            cash.iloc[bar] -= (units.diff().iloc[bar] @ self.price.iloc[bar]) 
            cash.iloc[bar] -= abs(units.diff().iloc[bar]) @ self.price.iloc[bar]*(0.001)
        return units, cash

class IncomeBacktest(BacktestStrategy):
    def __init__(self):
        super().__init__()
        
        # 엘엔에프(066970)와 에코프로비엠(247540)의 가격 데이터 가져오기
        price1 = fdr.DataReader('005930', '2019-01-01', '2023-05-10')['Close']
        price2 = fdr.DataReader('005935','2019-01-01', '2023-05-10')['Close']
        # 시초가 데이터 가져오기(전날 시그널이 뜨면 다음날 시초가에 매수/매도)
        o_price1 = fdr.DataReader('005930', '2019-01-01', '2023-05-10')['Open']
        o_price2 = fdr.DataReader('005935','2019-01-01', '2023-05-10')['Open']

        price =pd.concat([price1,price2],axis=1).dropna()
        price.columns = ['price1','price2']
        
        open_df = pd.concat([o_price1,o_price2],axis=1).dropna()
        open_df.columns = ['price1','price2']
    
        self.price = price
        # 인컴펀드 백테스트 
        
    def kf(self):
        data = self.price
        
        if not isinstance(data, np.ndarray):
            price_data = np.array(data)
        else:
            price_data = data

        # Set the measurement matrix
        measurement_matrix = np.array([[1, -1]])

        # Set the process noise and measurement noise covariances
        Q = 0.01 * np.eye(2)
        R = 1.0

        # Set the initial state estimate and covariance matrix
        x_hat = np.zeros((2, 1))
        P = np.eye(2)

        # Run the Kalman filter
        predicted_diff = np.zeros((len(price_data), 1))
        for t in range(len(price_data)):
            # Construct the measurement vector
            y = np.array([price_data[t, 0], price_data[t, 1]])
            measurement_vector = measurement_matrix @ y.reshape(-1, 1)

            # Prediction step
            x_hat_minus = x_hat
            P_minus = P + Q

            # Update step
            K = P_minus @ measurement_matrix.T @ np.linalg.inv(measurement_matrix @ P_minus @ measurement_matrix.T + R)
            x_hat = x_hat_minus + K @ (measurement_vector - measurement_matrix @ x_hat_minus)
            P = (np.eye(2) - K @ measurement_matrix) @ P_minus

            predicted_diff[t] = measurement_matrix @ x_hat

        return predicted_diff
        
    def kalman_signal(self,bar, predicted_diff,window, open_lim, close_lim):
        actual_diff =  np.array(self.price['price1']-self.price['price2']).reshape(-1,1)
        temp = (actual_diff-predicted_diff)[bar-window:bar]
        mean = temp.mean()
        std = temp.std()
        
        #print(mean+std, mean-std, (actual_diff-predicted_diff)[bar][0])
        if (actual_diff-predicted_diff)[bar][0] > mean + std :
            if (actual_diff-predicted_diff)[bar][0] > mean + close_lim * std :
                weight = [1 , 0]
            else:
                weight = [-1, 1]
        elif (actual_diff-predicted_diff)[bar][0] < mean - std :
            if (actual_diff-predicted_diff)[bar][0] < mean - close_lim * std :
                weight = [1,0.0]
            else:
                weight = [1, -1]
        else:
            weight = [1, 0.0]
        return weight
    
    
    def run_ls_backtest(self):
        total, units, cash = self.basic_tool()
        
        predicted_diff = self.kf()
        window = 20
        for bar in range(1, window+1):
            cash.iloc[bar] = cash.iloc[bar-1]
            total.iloc[bar] = total.iloc[bar-1]
            
        for bar in range(window+1,len(self.price)):
            #print( self.kalman_signal(bar, predicted_diff, window, 1, 2))
            cash.iloc[bar] = cash.iloc[bar-1]
            weight_t0 = self.kalman_signal(bar- 1, predicted_diff, window, 1, 2)
            weight_t1 = self.kalman_signal(bar - 2, predicted_diff, window, 1, 2)
            units, cash = self.order(bar, weight_t0 , weight_t1, units, cash)
            total.iloc[bar] += (self.price.iloc[bar] @ units.iloc[bar]) + cash.iloc[bar]
        return total, units, cash
    
if __name__ == "__main__":
    bt = IncomeBacktest()
    total, units, cash = bt.run_ls_backtest()
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import FinanceDataReader as fdr


class BacktestStrategy:
    def __init__(self):
        pass

    # 가져온 데이터들로 total, units, cash 데이터프레임을 만들어줌
    def basic_tool(self, initial_amount=10000000):
        price = self.price
        units = np.zeros((len(price), 2))
        total = pd.Series(index=price.index, dtype=float).fillna(0)  # 전체 총자산을 나타내는
        cash = pd.Series(index=price.index, dtype=float).fillna(0)

        total.iloc[0], cash.iloc[0] = initial_amount, initial_amount
        return total, units, cash

    def order(self, bar, weight_t0, weight_t1, units, cash):
        if np.array_equal(weight_t0, weight_t1):
            units[bar] = units[bar - 1]
        else:
            temp = (units[bar - 1] @ self.price.iloc[bar]) + cash[bar]
            units[bar] = (temp * np.array(weight_t0)) / self.price.iloc[bar]
            cash[bar] -= (np.abs(units[bar] - units[bar - 1]) @ self.price.iloc[bar]) * 0.001
            cash[bar] -= (units[bar] - units[bar - 1]) @ self.price.iloc[bar]
        return units, cash


class IncomeBacktest(BacktestStrategy):
    def __init__(self):
        super().__init__()

        # 엘엔에프(066970)와 에코프로비엠(247540)의 가격 데이터 가져오기
        price1 = fdr.DataReader('005930', '2019-01-01', '2023-05-10')['Close']
        price2 = fdr.DataReader('005935','2019-01-01', '2023-05-10')['Close']
        # 시초가 데이터 가져오기(전날 시그널이 뜨면 다음날 시초가에 매수/매도)
        o_price1 = fdr.DataReader('005930', '2019-01-01', '2023-05-10')['Open']
        o_price2 = fdr.DataReader('005935','2019-01-01', '2023-05-10')['Open']

        price =pd.concat([price1,price2],axis=1).dropna()
        price.columns = ['price1','price2']

        open_df = pd.concat([o_price1,o_price2],axis=1).dropna()
        open_df.columns = ['price1','price2']

        self.price = price

    def kf(self):
        data = self.price.values

        # Set the measurement matrix
        measurement_matrix = np.array([[1, -1]])

        # Set the process noise and measurement noise covariances
        Q = 0.01 * np.eye(2)
        R = 1.0

        # Set the initial state estimate and covariance matrix
        x_hat = np.zeros((2, 1))
        P = np.eye(2)

        # Run the Kalman filter
        predicted_diff = np.zeros((len(data), 1))
        for t in range(len(data)):
            # Construct the measurement vector
            y = np.array([data[t, 0], data[t, 1]])
            measurement_vector = measurement_matrix @ y.reshape(-1, 1)

            # Prediction step
            x_hat_minus = x_hat
            P_minus = P + Q

            # Update step
            K = P_minus @ measurement_matrix.T @ np.linalg.inv(measurement_matrix @ P_minus @ measurement_matrix.T + R)
            x_hat = x_hat_minus + K @ (measurement_vector - measurement_matrix @ x_hat_minus)
            P = (np.eye(2) - K @ measurement_matrix) @ P_minus

            predicted_diff[t] = measurement_matrix @ x_hat

        return predicted_diff

    def kalman_signal(self, bar, predicted_diff, window, open_lim, close_lim):
        actual_diff = self.price['price1'].values - self.price['price2'].values
        temp = (actual_diff - predicted_diff)[bar - window:bar]
        mean = np.mean(temp)
        std = np.std(temp)

        if (actual_diff - predicted_diff)[bar][0] > (mean + std)[0]:
            if (actual_diff - predicted_diff)[bar][0] > (mean + close_lim * std)[0]:
                weight = [1, 0]
            else:
                weight = [-1, 1]
        elif (actual_diff - predicted_diff)[bar][0] < (mean - std)[0]:
            if (actual_diff - predicted_diff)[bar][0] < (mean - close_lim * std)[0]:
                weight = [1, 0]
            else:
                weight = [1, -1]
        else:
            weight = [1, 0]

        return weight



    def run_ls_backtest(self):
        total, units, cash = self.basic_tool()

        predicted_diff = self.kf()
        window = 20

        for bar in range(1, window + 1):
            cash[bar] = cash[bar - 1]
            total[bar] = total[bar - 1]

        for bar in range(window + 1, len(self.price)):
            cash[bar] = cash[bar - 1]
            weight_t0 = self.kalman_signal(bar - 1, predicted_diff, window, 1, 2)
            weight_t1 = self.kalman_signal(bar - 2, predicted_diff, window, 1, 2)
            units, cash = self.order(bar, weight_t0, weight_t1, units, cash)
            total[bar] = (self.price.iloc[bar] @ units[bar]) + cash[bar]

        return total, units, cash

if __name__ == "__main__":
    bt = IncomeBacktest()
    total, units, cash = bt.run_ls_backtest()
```
